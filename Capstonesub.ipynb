## Data Description

The block of code below loads the data sets. It also loads the first five rows of the data this allows us to see the different columns and understand the differnt data types. The following output shows that there are 18 columns and states the different data types for each column. Additionally, this output counts the null elements in the dataset.

Acrroding to the output from the code that follows it can be seen that column names are as follows: 'date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot','floors', 'waterfront', 'view', 'condition', 'sqft_above','sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city','statezip', 'country'


Below is a detailed description of each feature:

####Date:
The date when the property was sold. This feature helps in understanding
the temporal trends in property prices.
####Price:
The sale price of the property in USD. This is the target variable we aim
to predict.
####Bedrooms:
The number of bedrooms in the property. Generally, properties with
more bedrooms tend to have higher prices.
####Bathrooms:
The number of bathrooms in the property. Similar to bedrooms,
more bathrooms can increase a property’s value.
####Sqft Living:
The size of the living area in square feet. Larger living areas are
typically associated with higher property values.
####Sqft Lot:
The size of the lot in square feet. Larger lots may increase a property’s
desirability and value.
####Floors:
The number of floors in the property. Properties with multiple floors may
offer more living space and appeal.
####Waterfront:
A binary indicator (1 if the property has a waterfront view, 0 other-
wise). Properties with waterfront views are often valued higher.
####View:
An index from 0 to 4 indicating the quality of the property’s view. Better
views are likely to enhance a property’s value.
####Condition:
An index from 1 to 5 rating the condition of the property. Properties
in better condition are typically worth more.
####Sqft Above:
The square footage of the property above the basement. This can
help isolate the value contribution of above-ground space.
####Sqft Basement:
The square footage of the basement. Basements may add value
depending on their usability.
####Yr Built:
The year the property was built. Older properties may have historical
value, while newer ones may offer modern amenities.
####Yr Renovated:
The year the property was last renovated. Recent renovations can
increase a property’s appeal and value.
####Street:
The street address of the property. This feature can be used to analyze
location-specific price trends.
####City:
The city where the property is located. Different cities have distinct market
dynamics.
####Statezip:
The state and zip code of the property. This feature provides regional
context for the property.
####Country:
The country where the property is located. While this dataset focuses
on properties in Australia, this feature is included for completeness.

Going through the data, we can classify the different data types present in the dataset:


*   Ordinal: Condition, View
*   Cardinal: Street, City, Statezip, and Country
*   Interval: Date
*   Ratio: Price, Bedrooms, Bathrooms, Sqft_living, Sqft_lot, FLoors , Waterfront, Sqft_above, Sqft_basement, Yr_built, Yr_renovated

##Summary Stats

The following block of code gives out the summary stats. It calculates the mean, mode, median, range, and std deviation for the specified columns. These columns are the columns that were classified under Ratio data type.
Additionally, the mode of the non ratio data types was calculated.

```python
# Summary Statistics
#display(df.describe())

# Create a list to store table data
data = []
datanonrat =[]

# Iterate through numerical columns and calculate statistics
for col in df.select_dtypes(include=["number"]).columns:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode().iloc[0]
    range_val = df[col].max() - df[col].min()
    std_val = df[col].std()

    # Append row to data list
    data.append([col, round(mean_val, 2), round(median_val, 2), mode_val, round(range_val, 2), round(std_val, 2)])

# Convert to Pandas DataFrame
summary_df = pd.DataFrame(data, columns=["Feature", "Mean", "Median", "Mode", "Range", "Std_deviation"])

# Display the DataFrame
display(summary_df)

# calculate mode for the non ratio data types
for col in df.select_dtypes(exclude=["number"]).columns:
  mode_val = df[col].mode().iloc[0]
  datanonrat.append([col,mode_val])

summarynonrat= pd.DataFrame(datanonrat, columns=["Feature", "Mode"])

display(summarynonrat)
```
